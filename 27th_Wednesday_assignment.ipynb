```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the data
data =  pd.read_excel("/content/January 2024 Community Policing Report.xls")

# Filter data for white and black individuals
white_data = data[data['Race'] == 'W']
black_data = data[data['Race'] == 'B']

# Count the number of searches conducted for white and black individuals
white_person_search_count = (white_data['Person Searched'] == 'Y').sum()
black_person_search_count = (black_data['Person Searched'] == 'Y').sum()

white_vehicle_search_count = (white_data['Vehicle Searched'] == 'Y').sum()
black_vehicle_search_count = (black_data['Vehicle Searched'] == 'Y').sum()

white_force_officer_count = (white_data['Force Used by Officer'] == 'Y').sum()
black_force_officer_count = (black_data['Force Used by Officer'] == 'Y').sum()

white_force_subject_count = (white_data['Force Used by Subject'] == 'Y').sum()
black_force_subject_count = (black_data['Force Used by Subject'] == 'Y').sum()

# Create a bar chart
search_types = ['Person Searched', 'Vehicle Searched', 'Force Used by Officer', 'Force Used by Subject']
white_search_counts = [white_person_search_count, white_vehicle_search_count, white_force_officer_count, white_force_subject_count]
black_search_counts = [black_person_search_count, black_vehicle_search_count, black_force_officer_count, black_force_subject_count]

bar_width = 0.35
index = range(len(search_types))

plt.bar(index, white_search_counts, bar_width, label='White', color='blue')
plt.bar([i + bar_width for i in index], black_search_counts, bar_width, label='Black', color='green')

plt.xlabel('Search Type')
plt.ylabel('Frequency')
plt.title('Comparison of Searches between White and Black Individuals')
plt.xticks([i + bar_width/2 for i in index], search_types, rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```
